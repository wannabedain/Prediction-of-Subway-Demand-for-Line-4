```python
import pandas as pd
import numpy as np
```

```python
weather=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/인공지능팀플/sw.csv", encoding='cp949')
weather=weather.drop(['지점'], axis=1)
weather
```

```python
weather.columns=['region', 'date', 'temperature', 'rainfall', 'wind_speed', 'humidity', 'snowfall']
```

```python
weather=weather.fillna(0)
weather
```

```python
weather = weather[weather['date'].str.contains('7:00|8:00|9:00')]
weather = weather[~weather['date'].str.contains('17:00|18:00|19:00')]
weather
```
